# Q1) The Probability of Shooting at a Target

A point is chosen at random from a disk of radius **10**. Let **A** be the event that the point lies within **1** unit of the boundary.

1. Model the experiment assuming the point is uniformly distributed with respect to area, and compute $P(A)$.

#1. Uniform Distribution Over Area

Given **A**  *be theevent that the point lies within 1 unit of bounary* so using this we can able to right the radius
* Outter limit $r = R$
* Inner limit $r = R-1$

so using those results we can able to right
 * $$A = \{ r : R - 1 \le r \le R \}$$


when point is uiformley distributed
  * Area of sample space : $\text{Area}(\Omega)$ = $\Pi * R^2$
  * Area of Event : $Area(A) = \pi R^2 - \pi (R - 1)^2$
  


   $$P(A) =   Area(A)/\text{Area}(\Omega)$$
   $$P(A) =  (\pi R^2 - \pi (R - 1)^2) /\pi * R^2$$
   $$P(A) = (2R-1)/R^2$$

   *given R = 10 so*  $$P(A) = 0.19$$
  


```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_disk_area(n_points=5000, radius=10):
    # 1. Uniformly distribute the angle theta between 0 and 2*pi
    theta = np.random.uniform(0, 2 * np.pi, n_points)

    # 2. Uniformly distribute with respect to AREA
    # We take the square root of a uniform variable to account for the
    # fact that area increases with the square of the radius.
    r = radius * np.sqrt(np.random.uniform(0, 1, n_points))

    # Convert polar coordinates (r, theta) to Cartesian (x, y)
    x = r * np.cos(theta)
    y = r * np.sin(theta)

    # Calculate how many points fall within 1 unit of the boundary (r >= 9)
    in_event_a = np.sum(r >= (radius - 1))
    prob_a = in_event_a / n_points

    return x, y, prob_a

# Execution and Visualization
n = 10000
x, y, p_a = simulate_disk_area(n)

plt.figure(figsize=(8, 8))
plt.scatter(x, y, s=1, alpha=0.5, label=f"Simulated P(A) ≈ {p_a:.4f}")

# Draw the disk boundary and the Event A boundary
circle_outer = plt.Circle((0, 0), 10, color='r', fill=False, lw=2, label="Boundary (r=10)")
circle_inner = plt.Circle((0, 0), 9, color='blue', fill=False, linestyle='--', label="Event A limit (r=9)")
plt.gca().add_patch(circle_outer)
plt.gca().add_patch(circle_inner)

plt.xlim(-11, 11)
plt.ylim(-11, 11)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Uniform Distribution Over Area")
plt.legend(loc='upper right')
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()

print(f"Theoretical P(A): 0.1900")
print(f"Simulated P(A):   {p_a:.4f}")
```

## 2. Uniform Distribution Over Radius and Angle

In this model, the distance $r$ is chosen uniformly from the interval $[0, R]$ independently of the angle. The probability is proportional to the length of the radial interval.

*   **Length of Sample Space:** $L(\Omega) = R$
*   **Length of Event A:** $L(A) = R - (R - 1) = 1$

**Probability Calculation:**
$$P(A) = \frac{L(A)}{L(\Omega)}$$
$$P(A) = \frac{R - (R - 1)}{R}$$
$$P(A) = \frac{1}{R}$$

*Given $R = 10$:*
$$P(A) = \frac{1}{10} = 0.1$$

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_disk_radial(n_points=5000, radius=10):
    # 1. Uniformly distribute the angle theta between 0 and 2*pi
    theta = np.random.uniform(0, 2 * np.pi, n_points)

    # 2. Uniformly distribute the RADIUS (Linear/Radial distribution)
    # Here we pick r directly from [0, 10] without a square root.
    r = np.random.uniform(0, radius, n_points)

    # Convert polar coordinates (r, theta) to Cartesian (x, y)
    x = r * np.cos(theta)
    y = r * np.sin(theta)

    # Calculate how many points fall within 1 unit of the boundary (r >= 9)
    in_event_a = np.sum(r >= (radius - 1))
    prob_a = in_event_a / n_points

    return x, y, prob_a

# Execution and Visualization
n = 10000
x, y, p_a = simulate_disk_radial(n)

plt.figure(figsize=(8, 8))
plt.scatter(x, y, s=1, alpha=0.5, color='orange', label=f"Simulated P(A) ≈ {p_a:.4f}")

# Draw the boundaries
circle_outer = plt.Circle((0, 0), 10, color='black', fill=False, lw=2)
circle_inner = plt.Circle((0, 0), 9, color='red', fill=False, linestyle='--', label="Event A limit (r=9)")
plt.gca().add_patch(circle_outer)
plt.gca().add_patch(circle_inner)

plt.xlim(-11, 11)
plt.ylim(-11, 11)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Uniform Distribution Over Radius")
plt.legend(loc='upper right')
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()

print(f"Theoretical P(A): 0.1000")
print(f"Simulated P(A):   {p_a:.4f}")
```

## 3. Why the Answers Differ

It is clearly visible that the density of the points differs between the two models. In the first model, the points are equally spread out across the entire surface; in the second, the points are more densely concentrated at the center.

The reason for this lies in how we define the distribution measure. In the first method, we define the points as being spread out equally by area. In the second method, we define the points as being spread out radially.

This means that if we draw the same imaginary ring (annulus) in both circles, the number of points within that ring in Circle 1 depends on the radius of the ring. However, in Circle 2, the number of points remains constant regardless of the radius.

***In conclusion: both models are correct within the context of their own specific measures***

#**Q2) Bertrand's Paradox**

* 1 .A chord of a circle is chosen “at random.” What is the probability that its length exceeds the side length of the inscribed equilateral triangle?

# 1. The "Random Endpoints" Method

In this model, we determine the probability by picking two random points on the circle's circumference to form a chord. To make the math easier to visualize, let's fix the first point **A** at one of the corners of our inscribed equilateral triangle. Since the second point **B** is chosen uniformly, it has an equal chance of landing anywhere along the full $360^\circ$ of the circle.

If we define **E** as the event where the resulting chord is longer than the side of the triangle, we can see from the geometry that point **B** must land on the specific arc located between the other two vertices. This "target zone" covers exactly $120^\circ$ of the circle.

Using these results, we can define our event space as:
$$E = \{ \theta : 120^\circ \le \theta \le 240^\circ \}$$

Because the points are distributed uniformly, the probability is simply the ratio of the favorable arc to the entire circumference. With a total sample space of $360^\circ$ and a success range of $120^\circ$, the calculation is straightforward:

$$P(E) = \frac{\text{Range}(E)}{\text{Range}(\Omega)} = \frac{120^\circ}{360^\circ}$$

*This gives us a final probability of:*
**$$P(E) = 1/3 \approx 0.33$$**

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from matplotlib.collections import LineCollection

def simulate_bertrand_rotation(n_samples=800):
    R = 1.0
    side_length = R * np.sqrt(3)  # Threshold length

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))

    # --- METHOD 1A: FIXED STARTING POINT ---
    # Logic: One point is static at the top (90 degrees)
    p1_fixed = np.array([0, R])
    theta2 = np.random.uniform(0, 2*np.pi, n_samples)
    x2_m1 = R * np.cos(theta2)
    y2_m1 = R * np.sin(theta2)

    lengths_m1 = np.sqrt((x2_m1 - p1_fixed[0])**2 + (y2_m1 - p1_fixed[1])**2)
    lines1 = [((p1_fixed[0], p1_fixed[1]), (x2_m1[i], y2_m1[i])) for i in range(n_samples)]
    prob1 = np.mean(lengths_m1 > side_length)

    # --- METHOD 1B: ROTATED STARTING POINTS (Whole Possibility) ---
    # Logic: Both points are random, effectively "rotating" the fixed point logic
    theta1_rand = np.random.uniform(0, 2*np.pi, n_samples)
    theta2_rand = np.random.uniform(0, 2*np.pi, n_samples)

    x1_m2, y1_m2 = R * np.cos(theta1_rand), R * np.sin(theta1_rand)
    x2_m2, y2_m2 = R * np.cos(theta2_rand), R * np.sin(theta2_rand)

    lengths_m2 = np.sqrt((x2_m2 - x1_m2)**2 + (y2_m2 - y1_m2)**2)
    lines2 = [((x1_m2[i], y1_m2[i]), (x2_m2[i], y2_m2[i])) for i in range(n_samples)]
    prob2 = np.mean(lengths_m2 > side_length)

    # --- SHARED PLOTTING RENDERER ---
    for ax, lines, lengths, prob, title, is_fixed in zip(
        [ax1, ax2],
        [lines1, lines2],
        [lengths_m1, lengths_m2],
        [prob1, prob2],
        ["The Logic of One Point", "The Whole Possibility (Rotation)"],
        [True, False]
    ):
        # Draw Circle
        circle = plt.Circle((0, 0), R, color='black', fill=False, lw=1.5)
        ax.add_artist(circle)

        # Draw Inscribed Triangle (Reference)
        tri_thetas = np.deg2rad([90, 210, 330, 90])
        ax.plot(R * np.cos(tri_thetas), R * np.sin(tri_thetas), color='green', lw=2, linestyle='--')

        # Draw Chords
        colors = ['orange' if l > side_length else 'royalblue' for l in lengths]
        alphas = [0.3 if l > side_length else 0.08 for l in lengths]
        lc = LineCollection(lines, colors=colors, alpha=alphas, lw=0.7)
        ax.add_collection(lc)

        # Draw Points
        if is_fixed:
            ax.scatter([p1_fixed[0]], [p1_fixed[1]], color='red', s=60, zorder=5, label='Fixed Start (P1)')
        else:
            ax.scatter(x1_m2[:50], y1_m2[:50], color='red', s=10, zorder=5, alpha=0.6, label='Random Starts')

        ax.set_title(f"{title}\nObserved Prob: {prob:.4f} (Target: 0.3333)", fontsize=14)
        ax.set_xlim(-1.1, 1.1)
        ax.set_ylim(-1.1, 1.1)
        ax.set_aspect('equal')
        ax.legend(loc='lower right')

    plt.suptitle("Bertrand's Paradox: Random Endpoint Method (Rotation Invariance)", fontsize=16, fontweight='bold')
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

simulate_bertrand_rotation(1000)
```

# 2. The "Random Radial Distance" Method

In this second approach, we look at the problem by picking a radius of the circle and then choosing a random point along that radius. The chord is then drawn as the unique line perpendicular to the radius at that exact spot.

To make it easier to see, we can pick any radius we like—the symmetry of the circle means the angle won't change the math. We then choose a distance **d** from the center, ranging anywhere from $0$ to the full radius $R$.

If we let **E** be the event where the chord is longer than the side of our inscribed triangle, geometry shows us that this only happens if the chord is close to the center. Specifically, the distance **d** has to be less than the triangle's apothem, which is exactly $R/2$.

So using those results, we can define our success condition as:
*   $$E = \{ d : 0 \le d \le R/2 \}$$

Since we are picking the point uniformly along the radius:
*   **Total possible space:** $\text{Range}(\Omega) = R$
*   **Favorable space:** $\text{Range}(E) = R/2$

   $$P(E) = \frac{\text{Range}(E)}{\text{Range}(\Omega)}$$
   $$P(E) = \frac{R/2}{R}$$

   *This gives us a final result of:*
   **$$P(E) = 1/2 = 0.5$$**

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from matplotlib.collections import LineCollection

def simulate_bertrand_method2_with_triangle(n_samples=800):
    R = 1.0
    side_length = R * np.sqrt(3)

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))

    # --- METHOD 2A: FIXED RADIUS LOGIC ---
    # We choose midpoints along the vertical radius (0, y)
    mid_y_m1 = np.random.uniform(-R, R, n_samples)
    dist_m1 = np.abs(mid_y_m1)
    h_m1 = np.sqrt(np.maximum(0, R**2 - dist_m1**2))

    # Endpoints are horizontal relative to the vertical radius
    lines1 = [((-h_m1[i], mid_y_m1[i]), (h_m1[i], mid_y_m1[i])) for i in range(n_samples)]
    lengths_m1 = 2 * h_m1
    prob1 = np.mean(lengths_m1 > side_length)

    # --- METHOD 2B: ALL RADII (ROTATION) ---
    theta_rand = np.random.uniform(0, 2*np.pi, n_samples)
    dist_m2 = np.random.uniform(0, R, n_samples)

    mx, my = dist_m2 * np.cos(theta_rand), dist_m2 * np.sin(theta_rand)
    tx, ty = -np.sin(theta_rand), np.cos(theta_rand)
    h_m2 = np.sqrt(np.maximum(0, R**2 - dist_m2**2))

    lines2 = [((mx[i] + tx[i]*h_m2[i], my[i] + ty[i]*h_m2[i]),
               (mx[i] - tx[i]*h_m2[i], my[i] - ty[i]*h_m2[i])) for i in range(n_samples)]
    lengths_m2 = 2 * h_m2
    prob2 = np.mean(lengths_m2 > side_length)

    # --- PLOTTING ---
    for ax, lines, lengths, prob, title, is_fixed in zip(
        [ax1, ax2],
        [lines1, lines2],
        [lengths_m1, lengths_m2],
        [prob1, prob2],
        ["Logic: Midpoints on Vertical Radius", "The Whole Possibility (All Radii)"],
        [True, False]
    ):
        # Outer Circle
        circle = plt.Circle((0, 0), R, color='black', fill=False, lw=1.5)
        ax.add_artist(circle)

        # Inscribed Triangle
        tri_t = np.deg2rad([90, 210, 330, 90])
        ax.plot(R * np.cos(tri_t), R * np.sin(tri_t), 'g--', lw=2, alpha=0.8, label='Inscribed Triangle')

        # Inner Boundary (R/2)
        inner_circle = plt.Circle((0, 0), R/2, color='gray', fill=False, linestyle=':', lw=1, label='R/2 Boundary')
        ax.add_artist(inner_circle)

        # Chords
        colors = ['orange' if l > side_length else 'royalblue' for l in lengths]
        alphas = [0.35 if l > side_length else 0.08 for l in lengths]
        lc = LineCollection(lines, colors=colors, alpha=alphas, lw=0.7)
        ax.add_collection(lc)

        if is_fixed:
            ax.plot([0, 0], [-R, R], color='red', lw=2, alpha=0.5, label='Fixed Radius')

        ax.set_title(f"{title}\nObserved Prob: {prob:.4f} (Target: 0.5000)", fontsize=13)
        ax.set_xlim(-1.1, 1.1)
        ax.set_ylim(-1.1, 1.1)
        ax.set_aspect('equal')
        ax.legend(loc='lower right', fontsize='small')

    plt.suptitle("Bertrand's Paradox Method 2: Random Radius", fontsize=16, fontweight='bold')
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

simulate_bertrand_method2_with_triangle(1000)
```

# 3. The "Random Midpoint" Method

In this third approach, we determine the probability by picking a random point anywhere inside the circle and treating it as the midpoint of a chord. Every point within the circle, except for the center itself, uniquely defines a single chord. To understand when a chord will be longer than the side of our inscribed equilateral triangle, we have to look at where that midpoint **P** is located.

Geometrically, the chord will only be long enough if its midpoint is relatively close to the center of the circle. Specifically, the point **P** must fall within a smaller, inner concentric circle that has a radius of exactly $R/2$. If the midpoint lands outside this inner circle, the resulting chord becomes too short.

If we let **E** be the event where the chord length is successful, our focus shifts to comparing the area of this "success zone" to the entire circle. The area of the full sample space $\Omega$ is $\pi R^2$, while the area of our event space **E** is the area of the smaller circle, which is $\pi (R/2)^2$ or $\frac{\pi R^2}{4}$.

Since the midpoint is chosen uniformly across the surface, we calculate the probability by taking the ratio of these two areas:

$$P(E) = \frac{\text{Area}(E)}{\text{Area}(\Omega)} = \frac{\pi R^2 / 4}{\pi R^2}$$

*Simplifying this ratio gives us our final result:*
**$$P(E) = 1/4 = 0.25$$**

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from matplotlib.collections import LineCollection

def simulate_bertrand_method3_final(n_samples=1000):
    R = 1.0
    side_length = R * np.sqrt(3)

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))

    # --- MATH LOGIC ---
    # To pick a random point uniformly in a 2D disk:
    # Radius must be sqrt(U) to account for area scaling
    r_mid = R * np.sqrt(np.random.uniform(0, 1, n_samples))
    theta_mid = np.random.uniform(0, 2*np.pi, n_samples)

    # Midpoint coordinates
    mx = r_mid * np.cos(theta_mid)
    my = r_mid * np.sin(theta_mid)

    # Chord direction (perpendicular to the radius to the midpoint)
    tx, ty = -np.sin(theta_mid), np.cos(theta_mid)
    h = np.sqrt(np.maximum(0, R**2 - r_mid**2)) # half-chord length

    # Chord endpoints
    x1, y1 = mx + tx*h, my + ty*h
    x2, y2 = mx - tx*h, my - ty*h

    lengths = 2 * h
    prob = np.mean(lengths > side_length)
    colors = ['orange' if l > side_length else 'royalblue' for l in lengths]

    # --- PLOTTING ---
    for ax, is_logic_plot in zip([ax1, ax2], [True, False]):
        # Draw main circle
        ax.add_artist(plt.Circle((0, 0), R, color='black', fill=False, lw=2))

        # Draw the inscribed triangle (Reference)
        # Vertices at 90, 210, 330 degrees
        tri_t = np.deg2rad([90, 210, 330, 90])
        ax.plot(R * np.cos(tri_t), R * np.sin(tri_t), 'g--', lw=2, label='Inscribed Triangle')

        # Draw the critical R/2 boundary
        inner_circle = plt.Circle((0, 0), R/2, color='red', fill=False, linestyle=':', lw=2, label='Critical Boundary (R/2)')
        ax.add_artist(inner_circle)

        if is_logic_plot:
            # Show midpoints
            ax.scatter(mx, my, c=colors, s=10, alpha=0.7)
            ax.set_title("Logic: Midpoints Distributed by Area\n(Orange points fall inside R/2 circle)", fontsize=13)
        else:
            # Show chords
            lines = [((x1[i], y1[i]), (x2[i], y2[i])) for i in range(n_samples)]
            alphas = [0.3 if l > side_length else 0.05 for l in lengths]
            lc = LineCollection(lines, colors=colors, alpha=alphas, lw=0.7)
            ax.add_collection(lc)
            ax.set_title(f"Whole Possibility: All Chords\nObserved Prob: {prob:.4f} (Target: 0.2500)", fontsize=13)

        # Formatting
        ax.set_xlim(-1.1, 1.1)
        ax.set_ylim(-1.1, 1.1)
        ax.set_aspect('equal')
        ax.legend(loc='lower right', fontsize='small')

    plt.suptitle("Bertrand's Paradox Method 3: Random Midpoint (Area Selection)", fontsize=16, fontweight='bold')
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

simulate_bertrand_method3_final(1200)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection

def simulate_bertrand_all_methods(n_samples=1000):
    R = 1.0
    side_length = R * np.sqrt(3)

    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 7))

    # --- METHOD 1: RANDOM ENDPOINTS (1/3) ---
    theta1 = np.random.uniform(0, 2*np.pi, n_samples)
    theta2 = np.random.uniform(0, 2*np.pi, n_samples)
    x1_m1, y1_m1 = R * np.cos(theta1), R * np.sin(theta1)
    x2_m1, y2_m1 = R * np.cos(theta2), R * np.sin(theta2)
    lengths_m1 = np.sqrt((x2_m1 - x1_m1)**2 + (y2_m1 - y1_m1)**2)
    lines1 = [((x1_m1[i], y1_m1[i]), (x2_m1[i], y2_m1[i])) for i in range(n_samples)]
    prob1 = np.mean(lengths_m1 > side_length)

    # --- METHOD 2: RANDOM RADIUS (1/2) ---
    theta_r = np.random.uniform(0, 2*np.pi, n_samples)
    dist_r = np.random.uniform(0, R, n_samples)
    mx2, my2 = dist_r * np.cos(theta_r), dist_r * np.sin(theta_r)
    tx2, ty2 = -np.sin(theta_r), np.cos(theta_r)
    h2 = np.sqrt(np.maximum(0, R**2 - dist_r**2))
    lines2 = [((mx2[i] + tx2[i]*h2[i], my2[i] + ty2[i]*h2[i]),
               (mx2[i] - tx2[i]*h2[i], my2[i] - ty2[i]*h2[i])) for i in range(n_samples)]
    lengths_m2 = 2 * h2
    prob2 = np.mean(lengths_m2 > side_length)

    # --- METHOD 3: RANDOM MIDPOINT (1/4) ---
    r_mid = R * np.sqrt(np.random.uniform(0, 1, n_samples))
    theta_mid = np.random.uniform(0, 2*np.pi, n_samples)
    mx3, my3 = r_mid * np.cos(theta_mid), r_mid * np.sin(theta_mid)
    tx3, ty3 = -np.sin(theta_mid), np.cos(theta_mid)
    h3 = np.sqrt(np.maximum(0, R**2 - r_mid**2))
    lines3 = [((mx3[i] + tx3[i]*h3[i], my3[i] + ty3[i]*h3[i]),
               (mx3[i] - tx3[i]*h3[i], my3[i] - ty3[i]*h3[i])) for i in range(n_samples)]
    lengths_m3 = 2 * h3
    prob3 = np.mean(lengths_m3 > side_length)

    # --- RENDER ALL PLOTS ---
    titles = ["Method 1: Random Endpoints", "Method 2: Random Radius", "Method 3: Random Midpoint"]
    probs = [prob1, prob2, prob3]
    all_lines = [lines1, lines2, lines3]
    all_lengths = [lengths_m1, lengths_m2, lengths_m3]
    targets = ["1/3 (0.33)", "1/2 (0.50)", "1/4 (0.25)"]

    for i, ax in enumerate([ax1, ax2, ax3]):
        # Circle & Triangle
        ax.add_artist(plt.Circle((0, 0), R, color='black', fill=False, lw=2))
        tri_t = np.deg2rad([90, 210, 330, 90])
        ax.plot(R * np.cos(tri_t), R * np.sin(tri_t), 'g--', lw=2, alpha=0.8)

        # Chords
        colors = ['orange' if l > side_length else 'royalblue' for l in all_lengths[i]]
        alphas = [0.3 if l > side_length else 0.04 for l in all_lengths[i]]
        lc = LineCollection(all_lines[i], colors=colors, alpha=alphas, lw=0.6)
        ax.add_collection(lc)

        ax.set_title(f"{titles[i]}\nProb: {probs[i]:.4f} (Target: {targets[i]})", fontsize=14)
        ax.set_xlim(-1.1, 1.1)
        ax.set_ylim(-1.1, 1.1)
        ax.set_aspect('equal')
        ax.axis('off')

    plt.suptitle("Bertrand's Paradox: Comparison of Three Chords Selection Methods", fontsize=18, fontweight='bold', y=1.05)
    plt.tight_layout()
    plt.show()

simulate_bertrand_all_methods(1500)
```

# Conclusion: The Bertrand Paradox

Looking at these three results—$1/3$, $1/2$, and $1/4$—we can clearly see that the "noise" pattern is different in each of the 3 cases. This is the major reason for the different answers: everything is based on our own perspective of how we define "random."

Whether we are looking at the endpoints, the radial distance, or the midpoint area, each method creates a different physical distribution of chords. So using those results, we can say that all three answers are correct depending on how we choose to measure the space.

It shows that "at random" depends entirely on the mechanism we use, and our perspective is what actually determines the final probability.

# **Q9) Information Available in a Discrete Random Variable**


### **1. Probability Space $(\Omega, \mathfrak{F}, P)$**
To define the probability space, we identify the sample space, the set of all possible events, and their associated likelihoods.

*   **Sample Space ($\Omega$):** $\Omega = \{G_1, G_2, D_1, D_2\}$.
*   **$\sigma$-algebra ($\mathfrak{F}$):** Since $\mathfrak{F} = \mathfrak{P}(\Omega)$, it is the power set containing all $2^4 = 16$ possible subsets of outcomes.
*   **Probability Measure ($P$):**
    *   $P(\{G_1\}) = 0.50$.
    *   $P(\{G_2\}) = 0.20$.
    *   $P(\{D_1\}) = 0.10$.
    *   $P(\{D_2\}) = 0.20$.

---

### **2. $\sigma$-algebra generated by $X$, $\sigma(X)$**
The random variable $X$ records only whether the item is **Good** ($0$) or **Defective** ($1$).

*   **Mapping:** $X(G_1) = 0, X(G_2) = 0, X(D_1) = 1, X(D_2) = 1$.
*   **Pre-images:**
    *   $X^{-1}(0) = \{G_1, G_2\}$.
    *   $X^{-1}(1) = \{D_1, D_2\}$.
*   **$\sigma(X)$:** The smallest $\sigma$-algebra making $X$ measurable.
    *   **$\sigma(X) = \{\emptyset, \Omega, \{G_1, G_2\}, \{D_1, D_2\}\}$**.

---

### **3. $\sigma$-algebra generated by $Y$, $\sigma(Y)$**
$Y$ records both quality and machine, assigning a unique value to every individual outcome.

*   **Mapping:** $Y(G_1)=1, Y(G_2)=2, Y(D_1)=3, Y(D_2)=4$.
*   **$\sigma(Y)$:** Since every singleton set is measurable, it generates the full power set.
    *   **$\sigma(Y) = \mathfrak{P}(\Omega)$**.

---

### **4. Proof: $\sigma(X) \subseteq \sigma(Y)$ and Information Content**
*   **Proof:** Every set in $\sigma(X)$ (e.g., $\{G_1, G_2\}$) is also a member of the power set $\sigma(Y)$.
*   **Explanation:** $Y$ contains **more information** than $X$. Knowing $Y$ allows you to deduce $X$ perfectly, but knowing $X$ (e.g., $X=1$) does not tell you the machine (whether $Y=3$ or $Y=4$).

# **10 Information Available in Continuous Random Varaibles**

## **1. Explicit Definition of Probability Space $(\Omega, \mathfrak{F}, P)$**
To define the probability space for this continuous bivariate system:

*   **Sample Space ($\Omega$):** $\Omega = \mathbb{R}^2$, representing all possible pairs of temperature ($t$) and pressure ($p$) deviations.
*   **$\sigma$-algebra ($\mathfrak{F}$):** $\mathfrak{F} = \mathcal{B}(\mathbb{R}^2)$, which is the Borel $\sigma$-algebra on $\mathbb{R}^2$ generated by open rectangles.
*   **Probability Measure ($P$):** The measure induced by the joint probability density function (PDF):
    $$f_{T,P}(t, p) = \frac{1}{4\pi} \exp\left( -\frac{t^2}{2} - \frac{p^2}{8} \right)$$

---

## **2. The $\sigma$-algebra generated by $X$, $\sigma(X)$**
The random variable $X$ records only the temperature deviation, defined as the projection $X(t, p) = t$.

*   **Formal Definition:** $\sigma(X) = \{ X^{-1}(B) : B \in \mathcal{B}(\mathbb{R}) \}$.
*   **Geometric Form:** $X^{-1}(B) = \{ (t, p) \in \mathbb{R}^2 : t \in B \} = B \times \mathbb{R}$.
*   **Interpretation:** $\sigma(X)$ consists of **"cylindrical" sets**. Visually, these are vertical strips in the $t-p$ plane. Because they extend infinitely in the $p$ direction, they contain zero information about pressure.

---

## **3. The $\sigma$-algebra generated by $Y$, $\sigma(Y)$**
$Y$ represents the full sensor state, defined as the identity map $Y(t, p) = (t, p)$.

*   **Definition:** Since $Y$ is the identity map on the sample space $\Omega$, the $\sigma$-algebra it generates is the finest possible for this space.
*   **Result:** $\sigma(Y) = \mathcal{B}(\mathbb{R}^2)$.

---

## **4. Relationship: $\sigma(X) \subseteq \sigma(Y)$ and Information Analysis**

### **Proof of Containment**
For any Borel set $B \subseteq \mathbb{R}$, the product set $B \times \mathbb{R}$ is a Borel set in $\mathbb{R}^2$. Thus, every set in $\sigma(X)$ is also a set in $\sigma(Y)$. Formally:
$$\sigma(X) \subseteq \sigma(Y)$$

### **Information Content Analysis**
*   **Information Dominance:** $Y$ contains strictly more information than $X$.
*   **Dependency:** Knowing $Y$ (the full state) implies knowing $X$ perfectly. However, knowing $X$ leaves the pressure component $P$ unknown.

---

## **5 & 6. Marginal Densities**

### **Marginal Density of $X$**
Since $T \sim N(0, 1)$, its marginal density is:
$$f_X(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}$$

### **Marginal Density of $Y$**
Since $Y$ records the full state $(T, P)$, its density is the joint density provided:
$$f_Y(t, p) = \frac{1}{4\pi} \exp\left( -\frac{t^2}{2} - \frac{p^2}{8} \right)$$

---

## **7. Probability Calculations**
Given that $T \sim N(0,1)$ and $P \sim N(0,4)$ are independent, we use the standard normal cumulative distribution function $\Phi(z)$:

| Event | Calculation | Result |
| :--- | :--- | :--- |
| **$P(X \le 0)$** | By symmetry of the Normal distribution around the mean $\mu=0$. | **0.5** |
| **$P(X > 1)$** | $1 - P(X \le 1) = 1 - \Phi(1) \approx 1 - 0.8413$ | **0.1587** |
| **$P(Y \in (-\infty, 0] \times \mathbb{R})$** | Equivalent to $P(T \le 0)$, as pressure is marginalized out. | **0.5** |
| **$P(Y \in [-1, 1] \times [-2, 2])$** | $P(-1 \le T \le 1) \cdot P(-2 \le P \le 2)$ <br> $= [\Phi(1) - \Phi(-1)] \cdot [\Phi(\frac{2}{2}) - \Phi(-\frac{2}{2})]$ <br> $= (0.6827) \cdot (0.6827)$ | **$\approx 0.4661$** |

---

## **11. Functional Relationship: $X = \pi_1 \circ Y$**
*   **Verification:** By definition, $X(t, p) = t$. The projection function $\pi_1$ applied to $Y$ yields $\pi_1(t, p) = t$.
*   **Measurability:** The projection map $\pi_1: \mathbb{R}^2 \to \mathbb{R}$ is continuous. In probability theory, any continuous function of a random variable is itself a random variable. This confirms $X$ is a **measurable function** of $Y$.

---

## **13 & 14. Differential Entropy ($h$)**
The differential entropy for a Normal distribution $N(\mu, \sigma^2)$ is $h = \frac{1}{2} \ln(2\pi e \sigma^2)$.

*   **$h(X)$:**
    $$h(N(0, 1)) = \frac{1}{2} \ln(2\pi e) \approx \mathbf{1.4189 \text{ nats}}$$
*   **$h(Y)$:**
    Since $T$ and $P$ are independent, $h(Y) = h(T) + h(P)$.
    $$h(P) = \frac{1}{2} \ln(2\pi e \cdot 4) = \frac{1}{2} \ln(8\pi e)$$
    $$h(Y) = \ln(4\pi e) \approx \mathbf{3.4383 \text{ nats}}$$

---

## **15 & 16. Conditional Entropy and Theory**

### **Calculations**
*   **$h(Y|X)$:** Because $P$ is independent of $T$, $h(Y|X) = h(P) = \mathbf{\frac{1}{2} \ln(8\pi e)}$.
*   **$h(X|Y)$:** Since $Y$ perfectly determines $X$, the uncertainty in $X$ given $Y$ is zero, which in the continuous limit leads to:
    $$h(X|Y) \to \mathbf{-\infty}$$
