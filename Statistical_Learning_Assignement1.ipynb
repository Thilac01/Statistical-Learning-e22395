# Q1) The Probability of Shooting at a Target

A point is chosen at random from a disk of radius **10**. Let **A** be the event that the point lies within **1** unit of the boundary.

1. Model the experiment assuming the point is uniformly distributed with respect to area, and compute $P(A)$.

when point is uiformley distributed
  area of the  sample space = $pi$ . $R^2$

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_target_practice(n_points=5000):
    # Set up the figure
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

    # --- Model 1: Uniform in Area ---
    # To get uniform area, we use the square root of the random variable
    r1 = 10 * np.sqrt(np.random.uniform(0, 1, n_points))
    theta1 = np.random.uniform(0, 2*np.pi, n_points)
    x1, y1 = r1 * np.cos(theta1), r1 * np.sin(theta1)

    # --- Model 2: Uniform in Radial Distance ---
    r2 = np.random.uniform(0, 10, n_points)
    theta2 = np.random.uniform(0, 2*np.pi, n_points)
    x2, y2 = r2 * np.cos(theta2), r2 * np.sin(theta2)

    # Plotting Logic
    for ax, x, y, title, prob in zip([ax1, ax2], [x1, x2], [y1, y2],
                                     ["Uniform by Area", "Uniform by Radial Distance"],
                                     [0.19, 0.10]):
        # Draw boundaries
        outer = plt.Circle((0, 0), 10, color='black', fill=False, lw=2)
        inner = plt.Circle((0, 0), 9, color='red', fill=False, linestyle='--', label='Boundary of A')
        ax.add_artist(outer)
        ax.add_artist(inner)

        # Scatter points
        ax.scatter(x, y, s=1, alpha=0.5, color='teal')

        # Formatting
        ax.set_xlim(-11, 11); ax.set_ylim(-11, 11)
        ax.set_aspect('equal')
        ax.set_title(f"{title}\nTheoretical P(A) = {prob}")
        ax.legend(loc='upper right')

    plt.tight_layout()
    plt.show()

simulate_target_practice()
```
