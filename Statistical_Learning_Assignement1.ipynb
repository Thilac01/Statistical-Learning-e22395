# Q1) The Probability of Shooting at a Target

A point is chosen at random from a disk of radius **10**. Let **A** be the event that the point lies within **1** unit of the boundary.

1. Model the experiment assuming the point is uniformly distributed with respect to area, and compute $P(A)$.

Given **A**  *be theevent that the point lies within 1 unit of bounary* so using this we can able to right the radius
* Outter limit $r = R$
* Inner limit $r = R-1$

so using those results we can able to right
 * $$A = \{ r : R - 1 \le r \le R \}$$


when point is uiformley distributed
  * Area of sample space : $\text{Area}(\Omega)$ = $\Pi * R^2$
  * Area of Event : $Area(A) = \pi R^2 - \pi (R - 1)^2$
  


   $$P(A) =   Area(A)/\text{Area}(\Omega)$$
   $$P(A) =  (\pi R^2 - \pi (R - 1)^2) /\pi * R^2$$
   $$P(A) = (2R-1)/R^2$$

   *given R = 10 so*  $$P(A) = 0.19$$
  


```python
import numpy as np
import matplotlib.pyplot as plt

# --- Configuration & Parameters ---
total_samples = 1000
R_outer = 10.0         # Outer boundary
R_inner_ring = 9.0     # Start of shaded region
np.random.seed(42)     # For repeatable results

# --- 1. Point Generation (Monte Carlo Sampling) ---
# To ensure uniform distribution by area, we use the sqrt of the random variable
# Area of a circle scales with r^2, so r = sqrt(Area/pi)
radii = R_outer * np.sqrt(np.random.rand(total_samples))
thetas = 2 * np.pi * np.random.rand(total_samples)

# Convert Polar to Cartesian coordinates
x_coords = radii * np.cos(thetas)
y_coords = radii * np.sin(thetas)

# --- 2. Statistical Analysis ---
# Identify points within the ring: 9 <= r <= 10
in_ring_mask = (radii >= R_inner_ring) & (radii <= R_outer)
count_in_ring = np.sum(in_ring_mask)
empirical_ratio = count_in_ring / total_samples

# Theoretical Area Calculation for verification
# Area_ring / Area_total = (pi*10^2 - pi*9^2) / (pi*10^2)
theoretical_ratio = (R_outer**2 - R_inner_ring**2) / (R_outer**2)

# --- 3. Professional Visualization ---
plt.style.use('seaborn-v0_8-muted') # Clean, academic style
fig, ax = plt.subplots(figsize=(9, 9))

# Draw the geometry
ring_patch = plt.Circle((0, 0), R_outer, color='lightgray', alpha=0.4, label='Shaded Ring Area')
ax.add_patch(ring_patch)
center_void = plt.Circle((0, 0), R_inner_ring, color='white')
ax.add_patch(center_void)

# Plot dot distributions
ax.scatter(x_coords[~in_ring_mask], y_coords[~in_ring_mask],
           s=12, c='#3498db', alpha=0.5, label='Inner Circle Dots')
ax.scatter(x_coords[in_ring_mask], y_coords[in_ring_mask],
           s=15, c='#e74c3c', alpha=0.8, label='Ring Area Dots')

# Plot Styling
ax.set_aspect('equal')
ax.set_xlim(-11, 11)
ax.set_ylim(-11, 11)
ax.set_title(f"Uniform Point Distribution Analysis\n"
             f"Points in Ring: {count_in_ring} | Total: {total_samples}", fontsize=14)
ax.legend(frameon=True, loc='upper right')
ax.grid(True, linestyle=':', alpha=0.6)

# Final Output Report
print("-" * 30)
print(f"GEOMETRIC ANALYSIS REPORT")
print("-" * 30)
print(f"Total Points Sampled : {total_samples}")
print(f"Points in Shaded Area: {count_in_ring}")
print(f"Empirical Ratio      : {empirical_ratio:.4f}")
print(f"Theoretical Ratio    : {theoretical_ratio:.4f}")
print(f"Deviation            : {abs(empirical_ratio - theoretical_ratio):.4f}")
print("-" * 30)

plt.show()
```
